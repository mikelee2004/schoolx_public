```python
class Tokenizer:
    def __init__(self):
        pass
    def tokenize(self, text):
        words = text.split()
        lfw = [word.lower() for word in words] #Вывод текста маленькими буквами
        cfw = [word.capitalize() for word in words] #Вывод текста, где все токены с большой буквы
        return lfw, cfw
    
tokenizer = Tokenizer()
text1 = "YA O4EN' HO4U POEST'! POETOMU ZAKAN4IVAYU ETU RABOTU I UHOZHU EATING!!!!!!"
tokens = tokenizer.tokenize(text1)

for token in tokens: 
    print(token)
```

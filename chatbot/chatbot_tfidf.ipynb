```python
import random
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

questions = ["Привет!",
            "Как дела?",
            "Что делаешь?",
            "Как погода?",
            "Как настроение?",
            "Как прошёл день?",
            "Какой ваш любимый цвет?",
            "Как тебя зовут?",
            "Сколько времени сейчас?",
            "Что такое гравитация?",
            "Что такое Интернет?",
            "Как приготовить пиццу?",
            "Что такое фотосинтез?",
            "Кто написал 'Войну и мир'?",
            "Как устроен Атлантический океан?",
            "Что такое глобальное потепление?",
            "Как работает солнечная энергия?",
            "Что такое демократия?",
            "Какие планеты входят в Солнечную систему?",

            ]
answers = ["Здравствуй!",
            "Хорошо, спасибо!",
            "Отвечаю на вопросы.",
            "Солнечная погода сегодня.",
            "Прекрасное настроение!",
            "Нормально!",
            "Я не имею предпочтений в цветах, так как я искусственный интеллект.",
            "Меня зовут Бахлыш",
            " Для уточнения времени, пожалуйста, укажите свой часовой пояс.",
            "Гравитация - это сила, притягивающая все объекты с массой друг к другу.",
            "Интернет - мировая система объединенных компьютерных сетей, обеспечивающая обмен информацией.",
            "Для приготовления пиццы вам понадобятся: тесто, соус, начинка, итальянские травы, сыр.",
            "Фотосинтез - процесс, при котором растения используют энергию света, чтобы превратить углекислый газ и воду в глюкозу и кислород.",
            "'Войну и мир' написал Лев Толстой.",
            "Атлантический океан является вторым по величине океаном и омывает берега Америки, Европы и Африки.",
            "Глобальное потепление - это увеличение средней температуры Земли из-за повышенного выброса парниковых газов.",
            "Солнечная энергия производится путем преобразования света и тепла от Солнца в электрическую энергию с помощью солнечных панелей.",
            "Демократия - это форма правления, в которой власть принадлежит народу или их выбранным представителям.",
            "В Солнечную систему входят восемь планет: Меркурий, Венера, Земля, Марс, Юпитер, Сатурн, Уран, Нептун.",
            ]

vectorizer = TfidfVectorizer(tokenizer=lambda x: x.split(" "), preprocessor=lambda x: x, lowercase=False)
X = vectorizer.fit_transform(questions)

def get_best_response(user_input):
    best_response_index = None
    Y = vectorizer.transform([user_input])
    similarities = cosine_similarity(Y, X).flatten()
    best_response_index = np.argmax(similarities)
    return answers[best_response_index]

while True:
    user_input = input("Вы: ")
    response = get_best_response(user_input)
    print("Бот: " + response)
```

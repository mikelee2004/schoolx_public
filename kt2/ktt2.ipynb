```python
class Matrix:
    def __init__(self, matrix):
        self.matrix = matrix

    def len(self):
        return len(self.matrix)
    
    def getShape(self):
        rows = len(self.matrix)
        columns = len(self.matrix[0]) if rows > 0 else 0
        return rows, columns
    
    #СЛОЖЕНИЕ МАТРИЦ

    def __add__(self, other):
        if isinstance(other, Matrix):
            if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                result = [[0] * len(self.matrix[0]) for _ in range(len(self.matrix))]
                for i in range(len(self.matrix)):
                    for n in range(len(self.matrix[0])):
                        result[i][n] = self.matrix[i][n] + other.matrix[i][n]
                return Matrix(result)
            else:
                raise ValueError('Matrices gotta be the same shape!')
        elif isinstance(other, (int, float)):
            result = [[0] * len(self.matrix[0]) for _ in range(len(self.matrix))]
            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] + other
            return Matrix(result)
        else: 
            raise TypeError("An argument is gotta be Matrix's object, int or float!")    
    #Вычитание

    def __sub__(self, other):
        if isinstance(other, Matrix):
            if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                result = [[0] * len(self.matrix[0]) for _ in range(len(self.matrix))]
                for i in range(len(self.matrix)):
                    for n in range(len(self.matrix[0])):
                        result[i][n] = self.matrix[i][n] - other.matrix[i][n]
                return Matrix(result)
            else:
                raise ValueError("Matrices are gotta be the same shape!")
        elif isinstance(other, (int, float)):
            result = [[0] * len(self.matrix[0]) for _ in range(len(self.matrix))]
            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] - other
            return Matrix(result)
        else:
            raise TypeError("An argument is gotta be Matrix's object or int/float!")


    #Деление
    def __truediv__(self, other):
        if isinstance(other, Matrix):
            
            if len(self.matrix) != len(other.matrix) or len(self.matrix[0]) != len(other.matrix[0]):
                raise ValueError('Matrices gotta be the same shape!')

            result = [[0 for _ in range(len(self.matrix[0]))] for _ in range(len(self.matrix))]

            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] / other.matrix[i][n]

            return Matrix(result)

        elif isinstance(other, (int, float)):
            result = [[0 for _ in range(len(self.matrix[0]))] for _ in range(len(self.matrix))]

            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] / other

            return Matrix(result)

        else:
            raise ValueError("An argument is gotta be Matrix's object or int/float")

    # Умножение 
    def __mul__(self, other):
        if isinstance(other, Matrix):
            if len(self.matrix) != len(other.matrix) or len(self.matrix[0]) != len(other.matrix[0]):
                raise ValueError("Matrices are gotta be the same shape!")
            result = [[0 for _ in range(len(self.matrix[0]))] for _ in range(len(self.matrix))]
            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] * other.matrix[i][n]
            return Matrix(result)
        elif isinstance(other, (int, float)):
            result = [[0 for _ in range(len(self.matrix[0]))] for _ in range(len(self.matrix))]
            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] * other
            return Matrix(result)
        else:
            raise ValueError("An argument is gotta be Matrix's object or int/float")
    
    # Перемножение матриц 

    def __matmul__(self, other):
        if isinstance(other, (int, float)):
            result = [[0] * len(self.matrix[0]) for _ in range(len(self.matrix))]
            for i in range(len(self.matrix)):
                for n in range(len(self.matrix[0])):
                    result[i][n] = self.matrix[i][n] * other
            return Matrix(result)   
        elif isinstance(other, Matrix):
            if len(self.matrix[0]) == len(other.matrix):
                result = [[0] * len(other.matrix[0]) for _ in range(len(self.matrix))]
                for i in range(len(self.matrix)):
                    for n in range(len(other.matrix[0])):
                        for k in range(len(other.matrix)):
                            result[i][n] += self.matrix[i][k] * other.matrix[k][n]
                return Matrix(result)
            else:
                raise ValueError("The number of columns in the first matrix does not match the number of rows in the second matrix!")
        else:
            raise TypeError("An argument is gotta be Matrix's object or int/float!")
        
    # Транспонирование матрицы
    
    def T(self):
        rows = len(self.matrix)
        cols = len(self.matrix[0])
    # Создание новой матрицы с транспонированными размерами
        result = [[0 for _ in range(rows)] for _ in range(cols)]

    # Заполнение элементов новой матрицы значениями из исходной матрицы
        for i in range(rows):
            for n in range(cols):
                result[n][i] = self.matrix[i][n]

        return Matrix(result)

    # Равенство 
    def __eq__(self, other):
        result = []
        for i in range(len(self.matrix)):
            row = []
            for n in range(len(self.matrix[0])):
                row.append((self.matrix[i][n]) == (other.matrix[i][n]))
            result.append(row)
        return Matrix(result)
    
    # Представление 
    def __repr__(self):
        return repr(self.matrix)
    

    def getVectors(self):
        if self.getShape()[0] == 4 and self.getShape()[1] == 4:
            v1, v2, v3, v4 = ([] for i in range(4))
            for i in self.matrix:
                v1.append(i[0])
                v2.append(i[1])
                v3.append(i[2])
                v4.append(i[3])
            return [v1,], [v2,], [v3,], [v4,]
        else:
            raise AttributeError("Program would not work untill it'll be correct. It works only with 4x4 shaped matrix!")
    
    @staticmethod
    def multiplyVectors(vr1, vr2):
        res = 0
        for i in range(len(vr2[0])):
            res += vr1[0][i] * vr2[0][i]
        return res
    
    def basis(self):
        if self.getShape()[0] == 4 and self.getShape()[1] == 4:
            v1,v2,v3,v4 = self.getVectors()
            f1 = v1

            f2 = (Matrix(v2) - (Matrix(f1)*(Matrix.multiplyVectors(v2,f1) / Matrix.multiplyVectors(f1,f1)))).matrix

            f3 = (Matrix(v3) - (Matrix(f1)*(Matrix.multiplyVectors(v3,f1) / Matrix.multiplyVectors(f1,f1))) -
                               (Matrix(f2)*(Matrix.multiplyVectors(v3,f2) / Matrix.multiplyVectors(f2,f2)))).matrix
            
            f4 = (Matrix(v4) - (Matrix(f1)*(Matrix.multiplyVectors(v4,f1) / Matrix.multiplyVectors(f1,f1))) -
                               (Matrix(f2)*(Matrix.multiplyVectors(v4,f2) / Matrix.multiplyVectors(f2,f2))) -
                               (Matrix(f3)*(Matrix.multiplyVectors(v4,f3) / Matrix.multiplyVectors(f3,f3)))).matrix
            return Matrix(f1), Matrix(f2), Matrix(f3), Matrix(f4)
        else:
            raise AttributeError("Program would not work untill it'll be correct. It works only with 4x4 shaped matrix!")
        
    def orthogonal(self):
        if self.getShape()[0] == 4 and self.getShape()[1] == 4:
            res = []
            for n in self.basis():
                res.append((n/Matrix.normCalc(n)).matrix)
            return(res)
        else:
            raise AttributeError("Program would not work untill it'll be correct. It works only with 4x4 shaped matrix!")
        
    @staticmethod
    def normCalc(vr):
        print(vr.matrix)
        result = vr.matrix[0][0]**2 + vr.matrix[0][1]**2 + vr.matrix[0][2]**2 + vr.matrix[0][3]**2
        return result**0.5
```

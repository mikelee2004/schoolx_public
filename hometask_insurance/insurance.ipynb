```python
Y = df.charges
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=5)
X_test
```

```python
#Рассчет среднеквадратичной ошибки
def mse(X,Y,vec):
    P = len(Y)
    ms = (1. / P) * np.sum((X.dot(vec) - Y) ** 2)
    return ms


#Реализует градиентный спуск для оптимизации модели линейной регрессии
def grad(X, Y, lr, iter):
    m = X.shape[0]  #Вычисление числа обучающих примеров
    X = np.hstack((np.ones((m, 1)), X))  #Добавление столбца из единиц к матрице признаков для учета свободного члена в модели
    ps = np.random.rand(X.shape[1])  #Инициализация параметров случайными значениями
    track = np.zeros((iter, 1))  #Инициализация массива для отслеживания среднеквадратичной ошибки (mse) на каждой итерации
    
    #Основной цикл градиентного спуска
    for i in range(iter):
        ps = ps - (2. / m * lr * np.dot(X.T, (np.dot(X, ps) - Y)))  #Обновление параметров с помощью алгоритма градиентного спуска
        track[i] = mse(X, Y, ps)

    return track, ps

#Совершение предсказаний с использованием оптимизированных параметров
def pred(X, ps):
    X = np.hstack((np.ones((X.shape[0], 1)), X))  #Добавление столбца из единиц к матрице признаков
    return np.dot(X, ps)  #Возвращение предсказанных значений
```

```python
from sklearn.metrics import r2_score
```

```python
track, weights = grad(X_train, Y_train, 0.0003, 100000)
# print(weights)

Y_pred = pred(X_test, weights)
# print(Y_pred)

print("R-squared score:", r2_score(Y_test, Y_pred))
```
